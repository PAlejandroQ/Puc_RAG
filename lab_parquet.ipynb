```python
def extraer_siglas_de_columnas(df: pd.DataFrame, columnas: list) -> pd.Series:
    """
    Extrae todas las siglas (2 o más letras mayúsculas) de las columnas
    especificadas en el DataFrame.

    Args:
        df: DataFrame de entrada.
        columnas: Lista de nombres de columnas a procesar.

    Returns:
        pd.Series: Serie con una lista de siglas únicas por fila.
    """
    # 1. Crear una Serie de texto combinando las columnas especificadas
    # Usamos .astype(str) para manejar valores NaN/None
    df['texto_combinado'] = df[columnas].astype(str).agg(' '.join, axis=1)

    # 2. Aplicar str.findall() con el patrón regex: ([A-Z]{2,})
    # Esto busca 2 o más mayúsculas consecutivas.
    siglas_encontradas = df['texto_combinado'].str.findall(r'([A-Z]{2,})')
    
    # 3. Eliminar duplicados dentro de cada lista y ordenar (opcional)
    siglas_unicas_por_fila = siglas_encontradas.apply(lambda x: sorted(list(set(x))))
    
    return siglas_unicas_por_fila

# Columnas donde buscar las siglas
columnas_a_procesar =['nomeVariavel', 'descricao']

# Ejecutar la función y añadir la nueva columna al DataFrame
read_csv_tags['siglas_extraidas'] = extraer_siglas_de_columnas(read_csv_tags.copy(), columnas_a_procesar)

print("\n--- DataFrame con Siglas Extraídas ---")
read_csv_tags[['nomeVariavel', 'descricao', 'siglas_extraidas']].to_csv('app/csv/tags_MRO3_siglas_to_ask_full.csv', index=False)
```
